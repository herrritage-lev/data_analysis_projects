```python
import pandas as pd
import numpy as np
import scipy.stats

df = pd.read_csv('Downloads/StarWars.csv', sep=';')
df_gr = df.groupby('gender')['fan'].value_counts().unstack()

dof = (df_gr.shape[0] - 1) * (df_gr.shape[1] - 1)

alpha = 0.05
crit_value = 3.84

n_tot = df_gr.sum().sum()

n_tot_yes = df_gr['да'].sum()
n_tot_no = df_gr['нет'].sum()

n_tot_w = df_gr.iloc[0].sum()
n_tot_m = df_gr.iloc[1].sum()

p_fan = n_tot_yes/n_tot
p_no_fan = n_tot_no/n_tot

n_fan_w = p_fan * n_tot_w
n_no_fan_w = p_no_fan * n_tot_w
n_fan_m = p_fan * n_tot_m
n_no_fan_m = p_no_fan * n_tot_m

df_gr_teor = np.array([[n_fan_w,n_no_fan_w],[n_fan_m,n_no_fan_m]])

chi_square = ((df_gr - df_gr_teor) ** 2/df_gr_teor).sum().sum()

scipy.stats.chi2_contingency(df_gr,correction=False)

chi_square > crit_value
```
